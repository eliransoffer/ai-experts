```python

def generate(cars: list[str]):
    client = genai.Client(
        api_key=userdata.get('GOOGLE_API_KEY'),
    )

    model = "gemini-2.0-flash"
    contents = [
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text=
                    f"""You are going to assist me with data extraction. Your task is to extract data about cars from the list you will be given.
                    List of cars: {cars}"""
                    ),
            ],
    )]
    generate_content_config = types.GenerateContentConfig(
        temperature=1,
        top_p=0.95,
        top_k=40,
        max_output_tokens=8192,
        response_mime_type="application/json",
        response_schema=genai.types.Schema(
            type = genai.types.Type.OBJECT,
            properties = {
                "cars": genai.types.Schema(
                    type = genai.types.Type.ARRAY,
                    items = genai.types.Schema(
                        type = genai.types.Type.OBJECT,
                        required = ["manufacturer", "model", "engine_volume_liters"],
                        properties = {
                            "manufacturer": genai.types.Schema(
                                type = genai.types.Type.STRING,
                            ),
                            "model": genai.types.Schema(
                                type = genai.types.Type.STRING,
                            ),
                            "engine_volume_liters": genai.types.Schema(
                                type = genai.types.Type.NUMBER,
                            ),
                        },
                    ),
                ),
            },
        ),
    )

    result = client.models.generate_content(
        model=model,
        contents=contents,
        config=generate_content_config,
    )
    # print(result.candidates[0].content.parts[0].text)
    return json.loads(result.candidates[0].content.parts[0].text)['cars']


```

```python
data = "https://storage.googleapis.com/edulabs-public-datasets/CAR%20DETAILS%20FROM%20CAR%20DEKHO.csv"
```

```python
import pandas as pd
df = pd.read_csv(data)
```

```python
res = generate(list(df['name'])[:10])
```

```python
# prompt: define chunks() generator

def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
all_cars = []
# generate in chunks of 100
for chunk in chunks(list(df['name'])[:150], 100):
    all_cars.extend(generate(chunk))
# generate(list(df['name']))
```
