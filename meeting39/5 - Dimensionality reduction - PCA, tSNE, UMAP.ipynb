```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
import matplotlib.pyplot as plt
import seaborn as sns
from plotly import express as px
```

# Prepare data

```python

# Load and scale the wine dataset
data = load_wine()
X = data.data
y = data.target
X_scaled = StandardScaler().fit_transform(X)
```

# PCA

```python
# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)


```

```python
X_scaled.shape
```

```python
X_pca.shape
```

```python
X_pca
```

```python
fig = px.scatter(x=X_pca[:, 0], y=X_pca[:, 1], color=y)
fig.update_yaxes(scaleanchor="x", scaleratio=1)
fig.show()
```

```python
# prompt: display amount of variance pca explain

pca.explained_variance_ratio_
```

#tSNE

```python
# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)
```

```python
fig = px.scatter(x=X_tsne[:, 0], y=X_tsne[:, 1], color=y)
fig.update_yaxes(scaleanchor="x", scaleratio=1)
fig.show()
```

# UMAP

```python
# Apply UMAP with a specific number of neighbors
umap_model = umap.UMAP(n_neighbors=25, random_state=42)
X_umap = umap_model.fit_transform(X_scaled)
```

```python
fig = px.scatter(x=X_umap[:, 0], y=X_umap[:, 1], color=y)
fig.update_yaxes(scaleanchor="x", scaleratio=1)
fig.show()
```

```python
# Apply UMAP with a specific number of neighbors
umap_model = umap.UMAP(n_neighbors=5, random_state=42)
X_umap = umap_model.fit_transform(X_scaled)
```

```python
fig = px.scatter(x=X_umap[:, 0], y=X_umap[:, 1], color=y)
fig.update_yaxes(scaleanchor="x", scaleratio=1)
fig.show()
```
