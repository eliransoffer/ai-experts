# Guidelines

You are going to train DNN on the Bank Marketing dataset.

1. Use `bank-additional-full.csv` as **train/val** dataset
2. Use `bank-additional.csv` as **test** dataset
3. Consider whether you want to split dataset to constant train + validation or use cross-validation.
4. Run short EDA on the data
5. Run simple preperocessing and cleaning on the data
6. Build DNN in iteration - start from simple DNN with 1 hidden layer and relatively small amount of neurons, and continue iterating by adding complexity no the DNN by adding more layers and / or neurons
7. Use strategies for reducing overfit if needed
8. Use BatchNormalization if needed
9. You might need to configure learning rate
10. Evaluate your training with tensorboard
11. Add callbacks if needed
